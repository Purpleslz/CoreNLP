```python
filename_list = ['all.txt',
                 'pleonastic.txt',
                 'quantrule.txt',
                 'partitiveRule.txt',
                 'bareNPRule.txt',
                 'percentsymbol.txt',
                 'percentandmoney.txt',
                 'isAdjectival.txt',
                 'stoplist.txt',
                 'nested.txt',
                ]
root_path = './'

word_field = 3
pos_field = 4
parse_field = 5
ner_field = 10

mention_file = '/home/zhfan/workspace/NLP/experiments/CoreNLP/output_auto/Tue-Aug-01-19-54-28-CST-2017.predicted.txt'
```

```python
mention_info = []
with open(mention_file, 'r') as f:
    doc_info = []
    sentence_info = []
    
    for line in f:
        line = line.strip()
        
        if line:
            if line.startswith("#begin"):
                doc_info = []
            elif line.startswith("#end"):
                mention_info.append(doc_info)
            else:
                sentence_info.append(line.split('\t'))
        else:
            # empty line
            doc_info.append(sentence_info)
            sentence_info = []
```
